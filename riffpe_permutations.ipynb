Set parameters:

- **n** is the base $Z_n$
- **l** is the length of a message

Now $m \in \mathcal{M} =  (Z_n)^l$

```python
n = 2
l = 2
key = b'\xefTT\xc89\xd0ap\xd7M\x97V\xd3\x82h\xeb'
tag = "abcdab"
```

```python
N = math.factorial(n ** l)
print(N)
```

Instantiate

Generate a random messsage from $\mathcal{M}$

```python
c = Riffpe(n, l, key, tag)

message = random_message(n, l)

print(message)
```

```python
def get_perm(c):
    alreadySeen = {}

    perm = ""
    for a in range(n):
        for b in range(n):
            enc = c.enc([a, b])
            dec = c.dec(enc)
#             print(a, b, "->", enc)
            perm += str(enc)
            if [a, b] != dec:
                print("wtf?", dec)

            if str(enc) in alreadySeen.keys():
                print("wtf?", alreadySeen[enc])
            else:
                alreadySeen[str(enc)] = f"{a}, {b}"
    return perm
```

```python
keyPossibility = "0123456789abcdef"

perms = {}

total = 0

for i in range(len(keyPossibility)):
    k = keyPossibility[i]
    for ii in range(len(keyPossibility)):
        kk = keyPossibility[ii]
        for iii in range(len(keyPossibility)):
            kkk = keyPossibility[iii]
            for iiii in range(len(keyPossibility)):
                kkkk = keyPossibility[iiii]
            #key = '\xef\xc9\xd0\xd7\x97\xd3\x82"\xe" + k
            key = f"112233445566{kkkk}{kkk}{kk}{k}"
            #print(len(key))
            c = Riffpe(n, l, key.encode("utf-8") , tag)
            perm = get_perm(c)
            if perm in perms.keys():
                perms[perm] += 1
            else:
                perms[perm] = 1

            total += 1
        
#print(len(perms))

tvd = 0

sep_min = 1
sep_max = 0

print("frequency:")

for perm in perms:
    freq = perms[perm] / total
    tvd += abs(freq - 1/N)
    print("\t", perm, freq)
    if freq < sep_min:
        sep_min = freq
    
    if freq > sep_max:
        sep_max = freq
    
    
    
print("TVD:", tvd)
print("freq_expected", 1/N)
print("freq_max", sep_max, 1 - N * sep_max)
print("freq_min", sep_min, 1 - N * sep_min)
```
